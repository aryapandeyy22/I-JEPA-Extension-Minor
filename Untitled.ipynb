```python
from PIL import Image
from torchvision import transforms

# Preprocess (same normalization as ImageNet)
transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=(0.485, 0.456, 0.406),
                         std=(0.229, 0.224, 0.225)),
])

data_dir = "data/tiny-imagenet-200"
dataset = datasets.ImageFolder(os.path.join(data_dir, "train"), transform=transform)

subset_size = 2000  # keep small for stability
tiny_subset, _ = torch.utils.data.random_split(dataset, [subset_size, len(dataset)-subset_size])
loader = torch.utils.data.DataLoader(tiny_subset, batch_size=2, shuffle=True)

print(f"Tiny-ImageNet loaded: {len(tiny_subset)} samples")
```

```python
class HierarchicalPredictor(nn.Module):
    def __init__(self, embed_dim=1280, hidden_dim=512):
        super().__init__()
        self.layer1 = nn.Sequential(
            nn.Linear(embed_dim, hidden_dim),
            nn.GELU(),
            nn.Linear(hidden_dim, embed_dim)
        )
        self.layer2 = nn.Sequential(
            nn.LayerNorm(embed_dim),
            nn.Linear(embed_dim, embed_dim)
        )

    def forward(self, x):
        out = self.layer1(x)
        return self.layer2(out) + out

predictor = HierarchicalPredictor().to(device)
print("Hierarchical Predictor initialized.")
```

```python
import torch
from torch import nn, optim
from torch.amp import autocast, GradScaler
import gc

# ==============================
#  Define cross-view loss
# ==============================
def cross_view_loss(p1, p2):
    # normalized cosine similarity
    p1 = nn.functional.normalize(p1, dim=-1)
    p2 = nn.functional.normalize(p2, dim=-1)
    return 2 - 2 * (p1 * p2).sum(dim=-1).mean()

# ==============================
#  Optimizer & scaler
# ==============================
optimizer = optim.AdamW(predictor.parameters(), lr=1e-4, weight_decay=1e-2)
scaler = GradScaler(enabled=False)   # MPS doesn't support AMP reliably

epochs = 5
print("Training initialized...")

for epoch in range(epochs):
    model.eval()        # encoder frozen (pretrained)
    predictor.train()   # only predictor learns
    epoch_loss = 0.0

    for step, (imgs, _) in enumerate(loader):
        imgs = imgs.to(device, non_blocking=True)

        # two random augmentations for cross-view consistency
        x1 = transforms.functional.hflip(imgs)
        x2 = transforms.functional.adjust_brightness(imgs, 1.2)

        optimizer.zero_grad(set_to_none=True)

        # hierarchical embedding prediction
        with torch.no_grad():
            e1 = model(x1)
            e2 = model(x2)

        # flatten if patch-wise
        if e1.dim() == 3:
            e1 = e1.mean(dim=1)
            e2 = e2.mean(dim=1)

        p1 = predictor(e1)
        p2 = predictor(e2)
        loss = cross_view_loss(p1, p2)

        loss.backward()
        torch.nn.utils.clip_grad_norm_(predictor.parameters(), 1.0)
        optimizer.step()

        epoch_loss += loss.item()

        if step % 50 == 0:
            print(f"Epoch {epoch+1}, Step {step}, Loss: {loss.item():.4f}")

        # free memory
        del imgs, x1, x2, e1, e2, p1, p2, loss
        if device.type == "mps":
            torch.mps.empty_cache()
        gc.collect()

    print(f"‚úÖ Epoch {epoch+1} completed | Avg Loss: {epoch_loss/(step+1):.4f}")

torch.save(predictor.state_dict(), "hierarchical_predictor.pth")
print("üéØ Training done. Saved hierarchical_predictor.pth")


```

```python
import torch, numpy as np, gc
from tqdm import tqdm
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

if os.path.exists("hierarchical_predictor.pth"):
    predictor.load_state_dict(torch.load("hierarchical_predictor.pth", map_location=device))
    predictor.eval()
    print("Loaded best hierarchical predictor for evaluation")


model.eval()
predictor.eval()
feats, labels = [], []

print("\nüîç Extracting embeddings ...")
with torch.no_grad():
    for imgs, labs in tqdm(loader):
        imgs = imgs.to(device)
        emb = predictor(model(imgs))
        if emb.dim() == 3:           # [B, N, D] ‚Üí [B, D]
            emb = emb.mean(dim=1)
        feats.append(emb.cpu())
        labels.append(labs)
        del imgs, labs, emb
        if device.type == "mps":
            torch.mps.empty_cache()
        gc.collect()

X = torch.cat(feats).numpy()
y = torch.cat(labels).numpy()
print(f"Feature matrix: {X.shape}, Labels: {y.shape}")

print("\n Training linear probe ...")
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

split = int(0.8 * len(X))
X_train, X_test = X_scaled[:split], X_scaled[split:]
y_train, y_test = y[:split], y[split:]

clf = LogisticRegression(
    max_iter=2000, solver="saga", multi_class="multinomial", n_jobs=-1
)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

acc = accuracy_score(y_test, y_pred)
print(f"Linear probe accuracy: {acc*100:.2f}%")
print("\nClassification report (macro avg):")
print(classification_report(y_test, y_pred, digits=3))

print("\n Running t-SNE visualization ...")
subset = min(2000, len(X))
tsne = TSNE(n_components=2, perplexity=30, init="pca", random_state=42)
X2 = tsne.fit_transform(X[:subset])

plt.figure(figsize=(9,7))
plt.scatter(X2[:,0], X2[:,1], c=y[:subset], cmap="tab20", s=6, alpha=0.8)
plt.title("t-SNE of Hierarchical Predictor Embeddings (Tiny-ImageNet subset)")
plt.xticks([]); plt.yticks([])
plt.tight_layout()
plt.savefig("tsne_embeddings.png", dpi=200)
plt.show()

print("\n Evaluation complete ‚Äî t-SNE saved as tsne_embeddings.png")
```
